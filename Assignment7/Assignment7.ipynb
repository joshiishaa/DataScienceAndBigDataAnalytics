```python
import pandas as pd
import string
import nltk
import seaborn as sns
!pip install python-docx
from docx import Document
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import warnings
warnings.filterwarnings('ignore')
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
from sklearn.naive_bayes import GaussianNB
import numpy as np
from nltk.stem import PorterStemmer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score, precision_score, recall_score, f1_score
from nltk.stem import WordNetLemmatizer
from string import digits
import re
```

```python
doc = Document('testdoc.docx')
text = ""
```

```python
for para in doc.paragraphs:
    print(para.text)
    text = text + para.text + " "
```

```python
def process(text):
    text = text.lower().replace('\n', ' ')\
          .replace('-', ' ').replace(':', ' ')\
          .replace(',', '').replace('"', '') \
          .replace("...", ".").replace("..", ".") \
          .replace("!", ".").replace("?", "") \
          .replace(";", ".").replace(":", " ").replace("%", " ").replace('\xa0',' ').replace('\d+', ' ')

    text = "".join(v for v in text if v not in string.punctuation).lower()
    text = text.encode("utf8").decode("ascii",'ignore')

    text = " ".join(text.split())
    return text
text = process(text)
text = re.sub(r'[0-9]+', '', text)
text
```

```python
text_vec = word_tokenize(text)
```

```python
#nltk.download('averaged_perceptron_tagger')
nltk.pos_tag(text_vec)
```

```python
new_text = []
stop_words = set(stopwords.words('english'))
for w in text_vec:
    if w not in stop_words:
        new_text.append(w)
```

```python
print(new_text)
```

```python
stem = PorterStemmer()
temp = []
for word in new_text:
        temp.append(stem.stem(word))
        
print(temp)
```

```python
temp3 = []
lemm = WordNetLemmatizer()
for word in temp:
        temp3.append(lemm.lemmatize(word))
```

```python
print(temp3)
```

```python
vocab = set()

for i in temp3:
    vocab.add(i)
```

```python
print(vocab)
len(vocab)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
str1 = ""

for i in temp3:
    str1 = str1 + i + " "
docs = [str1]
vectorizer = TfidfVectorizer()
vec = vectorizer.fit_transform(docs)
```
